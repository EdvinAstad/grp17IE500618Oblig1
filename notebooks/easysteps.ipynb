```python
df = housing.sample(frac=1, random_state=2)
train_df = df[:2344]
train_df = train_df.reset_index(drop=True)
test_df = df[2344:]
test_df = test_df.reset_index(drop=True)

train_df
```

```python
y_train = train_df["Pool Area"].to_numpy()
y_test = test_df["Pool Area"].to_numpy()

y_train.shape, y_test.shape

((2500,),(200,))
```

```python
test_df
```

```python
#Get error of baseline model that uses the average Pool Area in train_df

from sklearn.metrics import mean_absolute_error

average_median_pool_area = train_df["Pool Area"].mean()
baseline_model_test_predictions = [average_median_pool_area] * len(test_df)

mean_absolute_error(baseline_model_test_predictions, y_test)
```

```python
# Get X_train_3 which has "sale price","Lot Area" and
X_train_3 = train_df[['SalePrice', 'Year Remod/Add','Overall Qual']].to_numpy()

X_train_3.shape

#Choose the amount of training data

(800,3)
```

```python
# Get X_test_3 which has SalePrice, year remod/add and Overall quality.
# The test data training data.
# 2000 training data

X_test_3 = test_df[['SalePrice', 'Year Remod/Add','Overall Qual']].to_numpy()
X_test_3.shape

(2000, 3)
```

```python
# Get error of random forest training on just (X_train_3, y train) and testing on (X_test_3, y_test)
#Forrest training
# On average our model was of with 3
from sklearn.ensemble import RandomForestRegressor

forest_base = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_3, y_train)
forest_base_test_predictions = forest_base.predict(X_test_3)
mean_absolute_error(y_test,forest_base_test_predictions)
```

```python
#Train PCA (Linear Dimensionality Reduction) with 2-feature output

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca = pca.fit(X_train_3)

pca

PCA(n_components=2)
```

```python
# Use PCA to transform X_train_3 and X_test_3
X_train_pca = pca.transform(X_train_3)
X_test_pca = pca.transform(X_test_3)

X_train_pca.shape, X_test_pca.shape

((2000,2), (800,3))
```

```python
# Get error of Random Forest training on (X_train_pca, y_train) and testing on (X_test_pca, y_test)
# PCA did around close to the same result
forest_pca = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_pca, y_train)
forest_pca_test_predictions = forest_pca.predict(X_test_pca)
mean_absolute_error(y_test, forest_pca_test_predictions)
```
