```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as linreg
from sklearn.model_selection import train_test_split as ttsplit
from sklearn.metrics import mean_squared_error as mserror
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
```

```python
housing = pd.read_csv('AmesHousing.csv')
```

```python
housing.dtypes.value_counts()
```

```python
housing.dtypes
```

```python
overal_quality = housing['Overall Qual'].values.reshape(-1, 1)
sale_price = housing['SalePrice'].values
```

```python
model = linreg()
model.fit(overal_quality, sale_price)
slope = model.coef_[0]
intercept = model.intercept_

overal_quality_pred = model.predict(overal_quality)
```

```python
#Creates a graph based of sales price by overal quality
plt.scatter(overal_quality, sale_price, color='blue', label='Data')
plt.plot(overal_quality, overal_quality_pred, color='red', label='Regression Line')
plt.xlabel('overall quality')
plt.ylabel('price')
plt.title('price by overall quality')
plt.legend()
plt.show()
```
